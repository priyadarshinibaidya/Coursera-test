```python
#Importing necessary libraries
from scipy import linspace #Creates sequence of evenly space number
from scipy.integrate import solve_ivp #Solves an integral given an initial value in space or time
from scipy.integrate import odeint #Solves initial-value problem of ordinary differential equations
import matplotlib.pyplot as plt #For plotting graphs
import numpy as np #For extracting the ans in form of array

#Giving the constants by the user. Float used to even accaept decimals.
A= float(input("PLease give the value of A :"))
B= float(input("Please give the value of B :"))

#Defining our equations
def odes(z,t):
  x=z[0] #To get the output in the form of array
  y=z[1] #To get the output in the form of array
  dxdt= A + x*x*y - B*x -x
  dydt= B*x -x*x*y
  return [dxdt, dydt]

#Defining the steady state values value
x0, y0 = A, B/A
#Defiing the initial values with perturbation
z0=[x0+0.5,y0 +0.5]
#Defining the time interval
t=np.linspace(0,200,10000)
#Evaluating the itegral
z=odeint(odes,z0,t)

#Storing the result of matrix in x and y variables
x= z[:,0]
y= z[:,1]

#Plotting with desired size of graph
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(5)

plt.subplot(1,2,1) #For plotting both X/Y vs T and Y vs X grapgh simulataneously
plt.semilogy(t,x ,label="X") #Semilogy used to plot both X vs T and Y vs T together in the same plot
plt.semilogy(t,y ,label="Y")
plt.xlabel("Time")
plt.ylabel("X/Y")
plt.legend()

f = plt.figure()
f.set_figwidth(20)
f.set_figheight(5) 

plt.subplot(1,2,2)
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```
